## SVM - Support Vector Machine

```python
help(SVC)
```

```python
X = df.drop('Virus Present',axis=1)
y = df['Virus Present']
```

```python
model = SVC(kernel='linear', C=1)
model.fit(X, y)
```

```python
plot_svm_boundary(model, X, y)
```

# Hyper Parameters

## C

```python
model = SVC(kernel='linear', C=0.01)
model.fit(X, y)
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='linear', C=2)
model.fit(X, y)
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='linear', C=10)
model.fit(X, y)
plot_svm_boundary(model, X, y)
```

## Gamma

```python
model = SVC(kernel='rbf', C=0.01, gamma=0.5)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=0.1, gamma="scale")
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=1, gamma=0.5)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=1, gamma="scale")
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=10, gamma=0.5)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=10, gamma="scale")
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='sigmoid')
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

## Degree

```python
model = SVC(kernel='poly', C=1, degree=1)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='poly', C=1, degree=5)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

## Grid Search

```python
svm = SVC()
param_grid = {'C':[0.01,0.1,1,10,1000],
              'kernel':['linear','rbf','sigmoid','poly'],
              'gamma':["scale", "auto"],
              'degree':[1,2]}
grid = GridSearchCV(svm,param_grid)
```

```python
grid.fit(X,y)
```

```python
grid.best_score_
```
