## Elbow Method for Choosing Reasonable K Values

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
```

```python
test_error_rates = []
train_error_rates = []
for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train_scaled,y_train)
    y_pred_test = knn_model.predict(X_test_scaled)
    y_pred_train = knn_model.predict(X_train_scaled)
    test_error = 1 - accuracy_score(y_test,y_pred_test)
    test_error_rates.append(test_error)
    train_error = 1 - accuracy_score(y_train,y_pred_train)
    train_error_rates.append(train_error)
```

```python
plt.figure(figsize=(15,8))
plt.plot(range(1,30), test_error_rates, color='blue', linestyle='--', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')
```

```python
plt.figure(figsize=(15,8))
plt.plot(range(1,30), test_error_rates, color='blue', linestyle='--', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')
plt.hlines(y=0.051, xmin = 0, xmax = 30, colors= 'r', linestyles="--") #Bunlar ekledim
plt.hlines(y=0.057, xmin = 0, xmax = 30, colors= 'r', linestyles="--");
```

```python
plt.figure(figsize=(15,8))
plt.plot(range(1,30), test_error_rates, color='blue', linestyle='--', marker='o',
         markerfacecolor='red', markersize=10)
plt.plot(range(1,30), train_error_rates, color='green', linestyle='--', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')
plt.hlines(y=0.051, xmin = 0, xmax = 30, colors= 'r', linestyles="--") #Bunlar ekledim
plt.hlines(y=0.057, xmin = 0, xmax = 30, colors= 'r', linestyles="--");
```

## Scores by Various K Values

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    print("Test_Set\n")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set\n")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_scaled,y_train)
print('WITH K=1\n')
eval_metric(knn, X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=22)
knn.fit(X_train_scaled,y_train)
print('WITH K=22\n')
eval_metric(knn, X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=21)
knn.fit(X_train_scaled,y_train)
print('WITH K=21\n')
eval_metric(knn, X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=19)
knn.fit(X_train_scaled,y_train)
print('WITH K=19\n')
eval_metric(knn, X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train_scaled,y_train)
print('WITH K=9\n')
eval_metric(knn, X_train_scaled, y_train, X_test_scaled, y_test)
```

## Pipeline

What happens can be described as follows:

Step 1: The data are split into TRAINING data and TEST data according to ratio of train_test_split-train ve test olarak ayr

Step 2: the scaler is fitted on the TRAINING data - Train datasn fit et

Step 3: the scaler transforms TRAINING data - train datasn transform et

Step 4: the models are fitted/trained using the transformed TRAINING data

Step 5: the scaler is used to transform the TEST data

Step 6: the trained models predict using the transformed TEST data

pipe.fit(X_train, y_train)--> scaler.fit_transform(X_train) --> knn.fit(scaled_X_train, y_train)

```python
from sklearn.pipeline import Pipeline
```

```python
operations = [("scaler", StandardScaler()), ("knn", KNeighborsClassifier())]  
# Positional arguman soz konusu sirasi onemlidir. Aksi halde calismaz
# Ilkine Fit ve transform yapabilecek her seyi verebiliriz. 
# Ikincisine ise fit ve predict yapabilecegimiz her seyi verebiliriz.
# Tek bir model verebiliriz. Her modeli verip karsilastirma yapilamaz.
```

```python
Pipeline(steps=operations)
```

```python
pipe_model = Pipeline(steps=operations)
```

```python
pipe_model.fit(X_train, y_train)
```

pipe.predict(X_test) --> scaler.transform(X_test) --> knn.predict(scaled_X_test)

```python
y_pred = pipe_model.predict(X_test)
y_pred
```
